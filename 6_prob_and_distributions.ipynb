# Probability and Distribution

### Imports

```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

### Get some data

```python
# Continuouse
xs = np.linspace(-5, 10, 2000)
# Discrete
ks = np.arange(50)
```

## 1. Discrete PMFs
Bionomial distributions, Poisson distribution

```python
# D4 dice or 0.25 probability per outcome
pmf_binom = st.binom.pmf(ks, 50, 0.25)
plt.bar(ks, pmf_binom, label='Bionomial Example (dice)', alpha=0.8)

# 30 car crashes per night on average
pmf_poisson = st.poisson.pmf(ks, 30)
plt.bar(ks, pmf_poisson, label='Poisson Example (car crashse)', alpha=0.8)
plt.legend();
```

```python
# What is the chance of rolling 10 1's in a D4 dice if you roll 50 times?
print(st.binom.pmf(10, 50, 0.25))
```

```python
# If we have 30 crashes a night on average, what is the chance of getting 50?
print(st.poisson.pmf(50, 30))
```

## 2. Continuous PDFs
Uniform, normal, exponential, student-t, log-normal, skew-normal

```python
# Uniform (Lower and Upper Bounds)
pdf_uniform = st.uniform.pdf(xs, -4, 10)
plt.plot(xs, pdf_uniform, label='Uniform(-4,10)')

# Normal (Mean and St. Dev)
pdf_normal = st.norm.pdf(xs, 5, 2)
plt.plot(xs, pdf_normal, label='Normal(5,2)')

# Exponential (Rate and )
pdf_exponential = st.expon.pdf(xs, loc=-2, scale=2)
plt.plot(xs, pdf_exponential, label='Exponential(0.5)')

# Studnet-T (Degrees of freedon)
pdf_studentt = st.t.pdf(xs, 1)
plt.plot(xs, pdf_studentt, label='Student-T(1)')

# Log Normal (Scale parameter)
pdf_lognorm = st.lognorm.pdf(xs, 1)
plt.plot(xs, pdf_lognorm, label='Lognorm(1)')

# Skew Normal (Alpha parameter)
pdf_skewnorm = st.skewnorm.pdf(xs, -6)
plt.plot(xs, pdf_skewnorm, label='Skewnorm(-6)')

# Finish Chart
plt.legend()
plt.ylabel("Probability")
plt.xlabel("X");
```

### Note about `Loc` and `Scale`

```python
# In built transformation (Scipy takes normalization into account automatically)
plt.plot(xs, st.t.pdf(xs, 1, loc=4, scale=2), label='In-Built')
# Manual shifting of input data (Without normalization - hence the peak is higher)
plt.plot(xs, st.t.pdf((xs-4)/2, 1, loc=0, scale=1), label='Manual')
plt.legend();

```

Scipy has loc=0 and scale=1 by default

It is a linear transformation, but the normalization factor needs to then also be done manually, which scipy does automatically
